<h3> while문 만들기 </h3>

```python
prompt = """
    1. Add, 2. Sub, 3. Multiple, 4.Quit 숫자를 입력하세요
"""
```

```python
def add(i,j):
    return i + j
def sub(i,j):
    return i - j
def mul(i,j):
    return i * j
```

```python
number = 0
while number != 4: # 입력된 값이 4가 아니면 while문을 실행하고, 4이면 빠져나가라
    print(prompt)
    number = int(input())
    if(number == 1):
        print("input i")
        i = int(input())
        print("input j")
        j = int(input())
        print("add result : " , add(i,j))
    
```

## while문 강제로 빠져나가기

```python
coffee = 10
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다")
    coffee = coffee - 1
    print("남은 커피의 양은 %d 개 입니다.", coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다")
        break
```

## 문제1) 위의 코드에서 coffee를 1개씩 판매를 함(+1씩 증가) 하루 판매량(300잔)을 팔면 while문 종료

```python
sold_coffee = 0
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다")
    sold_coffee = sold_coffee + 1
    print("판매한 커피의 양은 %d 개 입니다.", sold_coffee)
    if sold_coffee == 300:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다")
        break
```

## 문제2) 5개를 판매했으니, 장부에 적는다.

```python
sold_coffee = 0
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다")
    sold_coffee = sold_coffee + 1
    
    if sold_coffee % 5 == 0:
        print("%d 개를 판매하였습니다" %sold_coffee)
        print('\n')
    else:
        print("판매한 커피의 양은 %d 개 입니다." %sold_coffee)
    
    if sold_coffee == 300:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다")
        break
```
